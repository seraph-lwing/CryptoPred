## Cleaning the text

```python
def clean_hashtags(string):
    replacements = [
    ("[",""),
    ("]",""),
    ("BITCOIN","BTC"),
    ("ETHEREUM","ETH"),
    ("TETHER","USDT"),
    ("BINANCE","BNB"),
    ("BINANCE COIN","BNB"),
    ("\r",""),
    ("'",""),
    (" ",""),
]
    #print(string)
    for old,new in replacements:
        string= string.strip().upper().replace(old,new).strip()
    return string
```

```python
clean_tweets["Hashtags\r"] = clean_tweets["Hashtags\r"].apply(clean_hashtags)
```

```python
hash_df = clean_tweets['Hashtags\r'].str.split(',', expand=True)
clean_tweets = clean_tweets.join(hash_df)
```

```python
clean_tweets
```

```python
tags=list(range(39))
clean_tweets[[0,1]].value_counts()
```

```python
final_count = Counter()
for col in tags:
    final_count = Counter(clean_tweets[col]) + final_count
print(type(final_count))
```

```python
sum(final_count.values())/len(final_count.values())
```

```python
def removeInfrequent(dictionary):
    d = dict()
    for key,values in  dictionary.items():
        if values > 50:
            d[key] = values
    return d
frequent_tags = removeInfrequent(final_count)
```

### creating image mask for hashtags

```python
tag_mask = np.array(Image.open("wordclouds/tag.png"))
tag_mask
```

```python
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
```

```python
transformed_tag_mask = np.ndarray((tag_mask.shape[0],tag_mask.shape[1]), np.int32)

for i in range(len(tag_mask)):
    transformed_tag_mask[i] = list(map(transform_format, tag_mask[i]))
plt.imshow(transformed_tag_mask)
```

```python
#tuples = [tuple(x) for x in df.values]
wordcloud = WordCloud(background_color='white',mask=transformed_tag_mask,contour_color='#1DA1F2',contour_width=1)
wordcloud.generate_from_frequencies(frequent_tags)
plt.figure()
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file("wordclouds/frequent_tags.png")
```

## frequency graph

```python
df = pd.DataFrame(frequent_tags.values(),frequent_tags.keys(),columns=['Frequent tags']).sort_values('Frequent tags')
plt.figure(figsize=(30,20))
sns.barplot(data=df.iloc[:20,:],x=df.index,y=df['Frequent tags'])
plt.xticks(rotation=90)
plt.savefig('most frequent tags.pdf',dpi=300)
```
